## Taxonomies for Genomic Data Visualization
### Challenges
- Genome Organization
    - Multiple chromosomes
    - Stored as linear sequence, but interations are non-linear:short-range, long range
- Pattern across Scales
    - Chromosome level to individual nucleotides(A,C,G,T)
- Many types of patterns along the genome(SNP, epigenomic peaks, genome ic rearragements, etc.)
- Everything is conected
    - Need to integrate different data types: sequence, expresssion levels, metabolites, phenotype information, etc
    - Need to load many different types of data into a sigle software
- Large space with sparse distribution of patterns across multiple scales
- Large data sets(in bytes)
- Privacy concerns(data access)

#### Gremlin

### Approach
- Only consider data that is visualized in the sequence context
- Treat genome as a coordinate system
- Seperate concerns:
    - What data types can be mapped to the genome?
    - How can the coordinate system be laid out?
    - How can the mapped data be encoded?
    - How can the mapped data be encoded?
    - What kind of tasks are users trying to address with genomic data?

#### Visual Taxonomy
- Coordinate System
    - Layout
        - linear
        - circular
        - space-filling
        - spatial(3-Dimensional) e.g. folding
    - Partition
        - Contiguous: end to end
        - Segregated: nextto each other, not connected
    - Abstraction
        - None
        - Partial e.g. only shows exon
        - Complete e.g. length of the exome, only gene basepair
    - Arrangement
        - Linear
            - Parallel
            - Serial
            - Orthogonal e.g. conservation
        - Circular
            - Parallel
            - Serial    
    - Tracks : 색깔은 많아서 재미있을지는 몰라도 특별한 것은 없다. 
        - Encoding
        - Argument
    - View configurations
        - Views: 한꺼번에 얼마나 많은 view를 볼 수 있는가? 
        - Scales: How many different scale you can look at a time point?
        - Foci: How many different loci you can look at a time point?
