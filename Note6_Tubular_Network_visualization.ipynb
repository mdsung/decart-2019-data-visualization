# Visualization of Trees and Networks
## Graph
- node and edges
- directed graph Vs. Undirected Graph

## Tree
- any number of chilren
- Only one parent

## Storing graphs and Trees
- Use a matrix for adjacency(edges explicit)
    - edgers are matrix cells
- Use a list with two fields per vertex(edges implicit)
    - incoming edges
    - outgoing edges
- Arrange Networks and Trees
    - Node link diagrams(대부분 이런식으로 그린다.)
    - Adjacency Matrix
    - Enclosure
        - containment marks
        
#### Node Link Diagram Approaches
- Where be drawn the nodes?
    - Triangular vertical layout
    - Radial layout
    - Rectangular horizontal layout
    - Bubble Tree layout
  
    - Circular layout
        - 모든 node를 circle에 배치하고 edge 연결
    - Force-directed layout
        - Hub를 눈으로 구분할 수 있다. cluster를 이루는 것도 확인할 수 있다. 
        - Hairballs
        - Ridiuclograms
    - minimizing edge crossing, if any draw perpenticular
- Where be drawn the deges?
